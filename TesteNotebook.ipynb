```python
import cv2
import math
import numpy as np
import array

white = [255,255,255]
black = [0,0,0]

mat = cv2.imread("reduzida2.png")

p = mat.shape
mRows, mCols, mType = mat.shape

def same(pixel1, pixel2):
    if len(pixel1)!= len(pixel2): return False
    for i in range(len(pixel1)):
        if pixel1[i]!=pixel2[i]: return False
    
    return True
```

```python
def sobel(x,y,c):
    ## gradient X ##
    gradx = 0-mat[x-1,y-1][c]-2*mat[x,y-1][c]-mat[x+1,y-1][c]+mat[x-1,y+1][c]+2*mat[x,y+1][c]+mat[x+1,y+1][c]
    ## gradient Y ##
    grady = 0-mat[x-1,y-1][c]-2*mat[x-1,y][c]-mat[x-1,y+1][c]+mat[x+1,y-1][c]+2*mat[x+1,y][c]+mat[x+1,y+1][c]
    grad = math.sqrt(gradx**2+grady**2)
    return grad

def medSobel(x,y):
    return int((sobel(x,y,0)+sobel(x,y,1)+sobel(x,y,2))/3)

#SOBEL
matSobel = mat.copy()
gradient = np.zeros((mRows,mCols,1), dtype=int)
menor = 255.0
maior = 0.0


for i in range(1, mRows-1, 1):
    for j in range(1, mCols-1, 1):
        g = medSobel(i,j)
        if menor > g:
            menor = g
        if maior < g:
            x = i
            y = j
            maior = g
        
        gradient[i,j] = g
        
for i in range(1, mRows-1, 1):
    for j in range(1, mCols-1, 1):
        grey = (gradient[i,j]-menor)*(255/(maior-menor))
#         if grey>200:
#             print("Alto")
        matSobel[i,j] = [grey, grey, grey]
        
print("{} {}".format(maior, menor))
print("{} {}".format(x, y))
#r = 1
for i in range(1, mRows-1, 1):
     for j in range(1, mCols-1, 1):
            if matSobel[i,j,0]> 50:
                matSobel[i,j] = [0, 255, 0]
        
# for i in range(1, mRows-1, 1):
#     for j in range(1, mCols-1, 1):
#         if same(matSobel[i,j],white):
#             print ("Branco")

cv2.imshow("Sobel", matSobel)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
def sobel(x,y):
    ## gradient X ##
    gradx = 0-mat[x-1,y-1][1]-2*mat[x,y-1][1]-mat[x+1,y-1][1]+mat[x-1,y+1][1]+2*mat[x,y+1][1]+mat[x+1,y+1][1]
    ## gradient Y ##
    grady = 0-mat[x-1,y-1][1]-2*mat[x-1,y][1]-mat[x-1,y+1][1]+mat[x+1,y-1][1]+2*mat[x+1,y][1]+mat[x+1,y+1][1]
    grad = math.sqrt(gradx**2+grady**2)
    return grad

#SOBEL
matSobel = mat.copy()
for i in range(1, mRows-1, 1):
    for j in range(1, mCols-1, 1):
        if sobel(i,j)>70:
            matSobel[i,j] = white
        else:
            matSobel[i,j] = black 
cv2.imshow("Sobel", matSobel)
```

```python
#Difference from the central pixel (Green Channel)
mRows, mCols, mType = mat.shape

def dif(x,y):
    diffG = 0
    for i in range(x-1, x+1, 1):
        for j in range(y-1, y+1, 1):
            diffG += math.fabs(int(mat[i,j][1])-int(mat[x,y][1]))
    return diffG  

matDiff = mat.copy()
for i in range(1, mRows-1, 1):
    for j in range(1, mCols-1, 1):
        if dif(i,j)>100:
            matDiff[i,j] = white
        else:
            matDiff[i,j] = black 
cv2.imshow("Difference", matDiff)
```

```python
# GREEN FILTER
output = mat.copy()
for i in range(mRows):
    for j in range(mCols):
        r = int(mat[i,j][0])
        g = int(mat[i,j][1])
        b = int(mat[i,j][2])
              
        if 2*g <= r+b or (r>200 and g>200 and b>200):
            output[i,j] = white
        else: 
            output[i,j] = black       
cv2.imshow("Fim", output)
```
